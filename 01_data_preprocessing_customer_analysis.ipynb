## Group-wise Imputation

```python
df['Education'].unique()
```

```python
df.groupby('Education')['Income'].median()
```

```python
# Function that returns median income based on education
def fill_income(row):
    if pd.isnull(row['Income']):
        return df[df['Education'] == row['Education']]['Income'].median()
    else:
        return row['Income']

df['Income'] = df.apply(fill_income, axis=1)

# Confirmation
df['Income'].isnull().sum()
```

## 2014 is the reference year

```python
reference_date = df['Dt_Customer'].max()
reference_date
```

```python
df['Age'] = df['Dt_Customer'].dt.year - df['Year_Birth']
df['Age'].describe().astype(int)
```

```python
df = df[df['Age'] <= 100]
```

```python
df['Age'].max()
```

```python
# First, import datetime just to be safe
from datetime import datetime

# Create a new column for Customer Tenure in days
df['Customer_Tenure'] = (reference_date - df['Dt_Customer']).dt.days
```

```python
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
df[['Kidhome', 'Teenhome', 'Total_Children']].head()
df['Total_Children'].value_counts()
```

```python
def age_bucket(age):
    if age <= 24:
        return "Gen Z"       # üíÖ Tiktok, digital-native, impulse buys
    elif age <= 34:
        return "Millennial"  # üß† Brand-aware, value-focused
    elif age <= 49:
        return "Gen X"       # üì¶ Practical, catalog bros
    elif age <= 64:
        return "Boomer"      # üßÇ Loyal, prefers phone/store
    else:
        return "Senior+"     # üë¥ Offline-first, gold buyers maybe

df['Age_Group'] = df['Age'].apply(age_bucket)
```

```python
df['Age_Group'].value_counts()
```

```python
product_columns = [
    'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'
]

df['Total_Spent'] = df[product_columns].sum(axis=1)
df['Total_Spent'].describe()
```

```python
df.drop(['Z_CostContact', 'Z_Revenue', 'ID'], axis=1, inplace=True)
```
